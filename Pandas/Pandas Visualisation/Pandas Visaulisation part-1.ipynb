```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/Users/tanishqsharma/Desktop/Python Projects/Pandas Visualisation/Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
#1. plot() function
df.plot()
#By Default it will give us line plot
```

```python
#We can also mentioned line this will also give us same result
df.plot(kind = 'line')
```

```python
#Plotting line graphs for each column seprately 
df.plot(kind = 'line', subplots = True)
```

```python
#We can add titles aswell on our graph
df.plot(kind = 'line', title = 'Ice cream Rating')
```

```python
#We can customize the labels of the graph as well
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
#2. Bar Charts
df.plot(kind = 'bar')
```

```python
# We can also make stacked bar chart
df.plot(kind = 'bar', stacked = True)
```

```python
#We can also apply arguments which we did in line chart
df.plot(kind = 'bar', title = 'Ice Cream Rating', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
#We can also make subplots in bar graph
df.plot(kind ='bar', subplots = True)
```

```python
#Using Every arguments on bar graph
df.plot(kind = 'bar', stacked = True, subplots = True, title = 'Ice Cream Rating', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
#We don't always have to do every single column, we can specify the column
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
#Making horizontal graph
df.plot.barh(stacked = True)
```

```python
#3. Scatter Plot
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
```

```python
#Specify the size of the scatter dots
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100)
```

```python
#Make size = 500:
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s =500)
```

```python
#We can also change dots colour
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 200 , c = 'red')
```
