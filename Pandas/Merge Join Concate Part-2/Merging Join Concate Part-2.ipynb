```python
#Join
import pandas as pd
```

```python
df1 = pd.read_csv(r'/Users/tanishqsharma/Desktop/Python Projects/Merge Join Concate Part-2/LOTR.csv')
df1
```

```python
df2 = pd.read_csv(r'/Users/tanishqsharma/Desktop/Python Projects/Merge Join Concate Part-2/LOTR 2.csv')
df2
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix = '_Left',rsuffix = '_Right',how = 'outer')
df4
```

```python
#Concat-> Concat is like putting one data frame on top of other.
pd.concat([df1,df2])
```

```python
#Concat with join (Inner join)
pd.concat([df1,df2], join = 'inner')
```

```python
#Concat with outer join
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```

```python
#Append function
df1.append(df2)
#Similar output as concat so it is recommend to use append
```
