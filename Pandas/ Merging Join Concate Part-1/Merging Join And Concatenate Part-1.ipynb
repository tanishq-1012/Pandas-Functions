```python
import pandas as pd
```

```python
df1 = pd.read_csv(r'/Users/tanishqsharma/Desktop/Python Projects/Merging Join and Concate/LOTR.csv')
df1
```

```python
df2 = pd.read_csv(r'/Users/tanishqsharma/Desktop/Python Projects/Merging Join and Concate/LOTR 2.csv')
df2
```

```python
#If nothing is specified it automatically performed inner join
df1.merge(df2)
```

```python
#Now if we specify inner it will give the same result
df1.merge(df2, how = 'inner')
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
#If we do it like this the we get the normal result
df1.merge(df2, how = 'inner', on = ['FellowshipID','FirstName'])
```

```python
#Outer Join
df1.merge(df2, how = 'outer')
#It Will give us NaN, it will do it in any value where could not find the join or could not match 
#values.
```

```python
#Left Join
df1.merge(df2, how = 'left')
#This is gonna take everything from left table data + overlap or common data from right table 
```

```python
#Right Join
df1.merge(df2, how = 'right')
#This is gonna take everything from right table data + overlap or common data from left table 
```

```python
#Cross Join
df1.merge(df2, how = 'cross')
```
